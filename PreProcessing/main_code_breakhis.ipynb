# Pre-Processing of the Images of the BreaKHis Dataset

The Breast Cancer Histopathological Image Classification (BreakHis) is  composed of 9,109 microscopic images of breast tumor tissue collected using different magnifying factors. The dataset is publicly available for download at: 
https://web.inf.ufpr.br/vri/databases/breast-cancer-histopathological-database-breakhis/

```python
import glob
from PIL import Image
import numpy as np
```

```python
#Define the directory of the download of the images
directory = "**"
```

```python
images = np.zeros((9109,128,128,3))
labels = []
count = 0

list_files = glob.glob(directory + "/benign/**/*.png", recursive=True)
for k in list_files:
    image = Image.open(k)
    new_image = image.resize((128, 128))
    new_image
    images[count] = np.asarray(new_image)
    labels.append((k.split("/")[-1]).split('.')[0])
    count = count+1
    
list_files = glob.glob(directory + "/malignant/**/*.png", recursive=True)
for k in list_files:
    image = Image.open(k)
    new_image = image.resize((128, 128))
    new_image
    images[count] = np.asarray(new_image)
    labels.append((k.split("/")[-1]).split('.')[0])
    count = count+1
```

```python
np.save("labels_BreaKHis.npy",labels)
np.save("images_BreaKHis.npy",np.uint8(images))
```
