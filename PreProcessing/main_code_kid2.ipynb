# Pre-Processing of the Images of the KID2 Dataset

The KID2 Dataset is composed of 2,371 wireless capsule endoscopy (WCE) images. The dataset is publicly available for download at: https://mdss.uth.gr/datasets/endoscopy/kid/

```python
import glob
from PIL import Image
import numpy as np
```

```python
#Define the directory of the download of the images
directory = "**"
```

```python
images = np.zeros((2371,128,128,3))
labels = []
classes = []
count = 0

list_files = glob.glob(directory + "/normal-small-bowel/*.png", recursive=True)
for k in list_files:
    image = Image.open(k)
    image = image.convert('RGB')
    new_image = image.resize((128, 128))
    new_image
    images[count] = np.asarray(new_image)
    labels.append(k.split("/")[-1])
    classes.append(0)
    count = count+1
    
list_files = glob.glob(directory + "/normal-stomach/*.png", recursive=True)
for k in list_files:
    image = Image.open(k)
    image = image.convert('RGB')
    new_image = image.resize((128, 128))
    new_image
    images[count] = np.asarray(new_image)
    labels.append(k.split("/")[-1])
    classes.append(1)
    count = count+1
    
list_files = glob.glob(directory + "/normal-esophagus/*.png", recursive=True)
for k in list_files:
    image = Image.open(k)
    image = image.convert('RGB')
    new_image = image.resize((128, 128))
    new_image
    images[count] = np.asarray(new_image)
    labels.append(k.split("/")[-1])
    classes.append(2)
    count = count+1
    
list_files = glob.glob(directory + "/normal-colon/*.png", recursive=True)
for k in list_files:
    image = Image.open(k)
    image = image.convert('RGB')
    new_image = image.resize((128, 128))
    new_image
    images[count] = np.asarray(new_image)
    labels.append(k.split("/")[-1])
    classes.append(3)
    count = count+1
    
list_files = glob.glob(directory + "/vascular/*.png", recursive=True)
for k in list_files:
    image = Image.open(k)
    image = image.convert('RGB')
    new_image = image.resize((128, 128))
    new_image
    images[count] = np.asarray(new_image)
    labels.append(k.split("/")[-1])
    classes.append(4)
    count = count+1

list_files = glob.glob(directory + "/inflammatory/*.png", recursive=True)
for k in list_files:
    image = Image.open(k)
    image = image.convert('RGB')
    new_image = image.resize((128, 128))
    new_image
    images[count] = np.asarray(new_image)
    labels.append(k.split("/")[-1])
    classes.append(5)
    count = count+1
    
list_files = glob.glob(directory + "/polypoids/*.png", recursive=True)
for k in list_files:
    image = Image.open(k)
    image = image.convert('RGB')
    new_image = image.resize((128, 128))
    new_image
    images[count] = np.asarray(new_image)
    labels.append(k.split("/")[-1])
    classes.append(6)
    count = count+1
    
list_files = glob.glob(directory + "/ampulla-of-vater/*.png", recursive=True)
for k in list_files:
    image = Image.open(k)
    image = image.convert('RGB')
    new_image = image.resize((128, 128))
    new_image
    images[count] = np.asarray(new_image)
    labels.append(k.split("/")[-1])
    classes.append(7)
    count = count+1
```

```python
np.save("labels_kid2.npy",labels)
np.save("classes_kid2.npy",classes)
np.save("images_kid2.npy",np.uint8(images))
```
